```python
import urllib3
import json
from urllib3 import Timeout
import os
```

## 4.2發送網絡請求
### 4.2.1 GET請求

```python
#PoolManager對象，調用request(method,url,fields=None,headers=None,**urlopen_kw)
url="http://httpbin.org/get"
http=urllib3.PoolManager()#創建連結池管理對象
r=http.request('GET',url)#發送GET請求
print(r.status)#print請求狀態碼
```

```python
#PoolManager對象 可以向多個服務器發送請求
urllib3.disable_warnings()#关闭SSL警告 SSL主要提供认证用户和服务器，确保数据发送到正确的客户机和服务器;加密数据以防止数据中途被窃取; 维护数据的完整性，确保数据在传输过程中不被改变。
jingdong_url="https://www.jd.com/"
Python_url="https://Python.org/"
baidu_url="https://www.baidu.com/"
http=urllib3.PoolManager()#創建連結池管理對象
r1=http.request('GET',jingdong_url)#發送GET請求
r2=http.request('GET',Python_url)#發送GET請求
r3=http.request('GET',baidu_url)#發送GET請求
print('JD Request status code:',r1.status)
print('Python Request status code:',r2.status)
print('BaiDu Request status code:',r3.status)
```

### 4.2.2POST请求

```python
#使用 request()方法实现POST请求 methane fields
urllib3.disable_warnings()#关闭SSL警告
url="http://httpbin.org/post"
params={'name':'Jack','region':'台湾','age':30}#定义字典类型的请求参数
http=urllib3.PoolManager()#創建連結池管理對象
r=http.request('POST',url,fields=params)#发送POST请求
print('The result is',r.data.decode('utf-8'))
print(r.data.decode('unicode_escape'))
```

### 4.2.3重试请求

```python
#retries重试请求 总会因为某种原因调用失败，这个时候我们可以通过重试的方式，来重新请求接口。
urllib3.disable_warnings()#关闭SSL警告
url="http://httpbin.org/get"
http=urllib3.PoolManager()#創建連結池管理對象
r=http.request('GET',url)#發送GET請求，默認重識請求
r1=http.request('GET',url,retries=5)#發送GET請求，5times重識請求
r2=http.request('GET',url,retries=False)#發送GET請求，close重識請求
print('默認重識請求次數：',r.retries.total)
print('設置重識請求次數：',r1.retries.total)
print('關閉重識請求次數：',r2.retries.total)
```

### 4.2.4处理响应内容

```python
#获取相应头 HTTPResponse对象 info()方法 type=dict字典 for循环遍历
urllib3.disable_warnings()#关闭SSL警告
url="http://httpbin.org/get"
http=urllib3.PoolManager()#創建連結池管理對象
r=http.request('GET',url)#發送GET請求，默認重識請求
response_header=r.info()#获取header
for key in response_header.keys(): #for循环遍历header information
    print(key,':',response_header.get(key))
print(response_header.keys())
```

```python
#处理服务器返回的JSON信息 JSON(JavaScript Object Notation，JS 对象标记)在互联网软件前端与后台进行消息交互的过程中，需要有一种标准的数据交换格式供前后端采用。
urllib3.disable_warnings()#关闭SSL警告
url="http://httpbin.org/post"
params={'name':'Jack','region':'台湾','age':30}#定义字典类型的请求参数
http=urllib3.PoolManager()#創建連結池管理對象
r=http.request('POST',url,fields=params)#发送POST请求
j=json.loads(r.data.decode('unicode_escape'))
print('The type of Data:',type(j))
print('Getting the form of the Data ',j.get('form'))
print('Getting the Region information',j.get('form').get('region'))
```

```python
#二进制数据 图片的二进制数据 可以使用open()函数
urllib3.disable_warnings()#关闭SSL警告
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:77.0) Gecko/20100101 Firefox/77.0'}
url="https://img.icons8.com/emoji/48/000000/-emoji-admission.png"
http=urllib3.PoolManager(headers=headers)#創建連結池管理對象
r=http.request('GET',url)#發送GET請求，默認重識請求
print(r.data)#print 二进制数据
f=open('landscape.jpg','wb+') #create open object
f.write(r.data)#write the data
f.close()#close
```

### 4.3.1设置header

```python
urllib3.disable_warnings()#关闭SSL警告
url="http://httpbin.org/get"
headers={'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:92.0) Gecko/20100101 Firefox/92.0'}
http=urllib3.PoolManager()#創建連結池管理對象
r=http.request('GET',url,headers=headers)#發送GET請求
print(r.data.decode('utf-8'))
```

```python
### 4.3.2设置超时
#作用：是为了防止url不可访问，或者响应速度太慢而造成的时间浪费。
#比如，你要爬取1000个网站，如果有100个需要30s才能返回数据，
#你等待他们返回的话就需要3000s了，如果你设置10s超时，那么就能知道最长需要多久1000个可以爬完
'''
try:
<语句>        #运行别的代码
except <名字>：
<语句>        #如果在try部份引发了'name'异常
except <名字>，<数据>:
<语句>        #如果引发了'name'异常，获得附加的数据
else:
<语句>        #如果没有异常发生
'''
urllib3.disable_warnings()               # 关闭ssl警告
baidu_url = 'https://www.baidu.com/'    # 百度超时请求测试地址
python_url = 'https://www.python.org/'  # Python超时请求测试地址
http = urllib3.PoolManager()                   # 创建连接池管理对象
try:
    r = http.request('GET',baidu_url,timeout=0.01)# 发送GET请求，并设置超时时间为0.01秒
except  Exception as error:
    print('百度超时：',error)
http2 = urllib3.PoolManager(timeout=0.1)  # 创建连接池管理对象,并设置超时时间为0.1秒
try:
    r = http2.request('GET', python_url)  # 发送GET请求
except  Exception as error:
    print('Python超时：',error)
#超时异常
```

```python
urllib3.disable_warnings()               # 关闭ssl警告
timeout=Timeout(connect=0.5,read=0.1) #设置连接0.5秒，读取0.1秒
http=urllib3.PoolManager(timeout=timeout)
r=http.request('GET','https://www.python.org/')
print(r.data.decode('utf-8'))
```

```python
timeout=Timeout(connect=0.5,read=0.1) #设置连接0.5秒，读取0.1秒
http=urllib3.PoolManager()
r=http.request('GET','https://www.python.org/',timeout=timeout)
print(r.data.decode('utf-8'))
```

### 4.3.3设置代理

```python
#1.proxy_url,代理IP，2.header 为了模拟游览器请求，避免后台服务器发现
url = "http://httpbin.org/ip"            # 代理IP请求测试地址
# 定义火狐浏览器请求头信息
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:77.0) Gecko/20100101 Firefox/77.0'}
# 创建代理管理对象
proxy = urllib3.ProxyManager('http://20.47.108.204:8888',headers = headers)
r = proxy.request('get',url,timeout=2.0)  # 发送请求
print(r.data.decode())                    # 打印返回结果
```

```python
with open('test.txt',encoding='GBK') as f:    # 打开文本文件
  data = f.read()
http = urllib3.PoolManager()    # 创建连接池管理对象
# 发送网络请求
r = http.request( 'POST','http://httpbin.org/post',fields={'filefield': ('example.txt', data),})
files = json.loads(r.data.decode('GBK'))['files']  # 获取上传文件内容
print(files)
```
